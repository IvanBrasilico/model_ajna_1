```python
with open('sizes_recinto2020311-3.pickle', 'rb') as handle:
    sizes_recinto = pickle.load(handle)
```

```python
sizes_listed = []
for recinto, sizes in sizes_recinto.items():
    for size in sizes:
        cols, rows = size
        sizes_listed.append((recinto, cols, rows))
    
```

```python
df = pd.DataFrame(sizes_listed, columns = ['recinto', 'cols', 'rows'])
df['ratio'] = df['cols'] / df['rows']
```
